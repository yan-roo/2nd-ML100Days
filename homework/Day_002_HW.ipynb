#### 截取部分資料

```python
select = pd.DataFrame(app_train.SK_ID_CURR)
select['CONTRACT_TYPE'] = pd.DataFrame(app_train.NAME_CONTRACT_TYPE)
#app_train.iloc[:10, :]

temp = app_train['NAME_CONTRACT_TYPE'].str.split(" ",  n=1, expand=True)
select['Loans_Type'] = temp[0]

select['Gender'] = pd.DataFrame(app_train.CODE_GENDER).replace(['M','F'], ['Male','Female'])

temp = app_train[['CNT_CHILDREN']].astype('str')
select['Famliy_withChildren'] = app_train['NAME_FAMILY_STATUS'].str.cat(temp, sep=' - ')
select
```
