* 作業2：試著想想看, 非監督學習是否有可能使用評價函數 (Metric) 來鑑別好壞呢?  
(Hint : 可以分為 "有目標值" 與 "無目標值" 兩個方向思考)<br><br>

    - __有目標值的分群:  如果資料有目標值，只是先忽略目標值做非監督學習，則只要微調後，就可以使用原本監督的測量函數評估準確性。__<br>
    分類是給定目標數量項目進行分發分類, 分群是利用演算法自動將相近的群體集合標示成同一群; 有目標值的分群則是以分群的方式計算, 最後用Label評估, 類似於分類。 <br><br>
    - __無目標值的分群:  但通常沒有目標值/目標值非常少才會用非監督模型，這種情況下，只能使用資料本身的分布資訊，來做模型的評估。__
    
<br><br>
[Performance metrics to evaluate unsupervised learning
Ask](https://stats.stackexchange.com/questions/79028/performance-metrics-to-evaluate-unsupervised-learning)<br>
[Evaluation Metrics for Unsupervised Learning Algorithms](https://arxiv.org/abs/1905.05667)<br>
論文中提及有兩種評估非監督式學習好壞的方式
1. 內部驗證函數(Internal Validation Metrics) 不需要外部資訊, 主要利用統計的方式測量群與群間的"聚合"與"離散"程度(例如:SSE[Sum of Squared Errors] 歐氏幾何距離), 距離矩陣(Proximity Matrix), 階層分群(Hierarchical Methods)。
2. 外部驗證函數(External Validation Metrics) 利用外部資訊, 其他演算法產生的結果相互比較(Matching Set與分類結果比較), 資訊增益(純度Entropy)。
