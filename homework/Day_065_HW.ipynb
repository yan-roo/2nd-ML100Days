### 參數想法
- LearningRate隨著BatchSize調整; BatchSize愈大代表每個批次裡面的資料量愈多, 使用高學習率比較能在500次內到達最佳解, 反之亦然。<br><br>
- 激活函數(Activation)<br>
Sigmoid有計算e上的負擔,以及梯度消失(轉換成-1~1之間)與非零中心(鋸齒狀移動)<br>
Tanh則仍有梯度消失的問題<br>
ReLU改善了以上三種問題, 但有梯度死亡的可能(離群值大變化造成參數調整不回來)<br>
所以採用ReLU<br><br>
- 正規化(Regularization)則是完全不採用, 因為沒有Noise雜訊的存在
